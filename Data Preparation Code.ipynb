Let's drop all rows where there is GCP or any other parenthesis like (Jones et al. 2023)

```python
github = github[~github['country'].str.contains('\(|\)')]
```

```python
github[github['iso_code'].isna()]['country'].unique()
```

Everything except Kosovo and Timor (which are countries, so they have a specific iso code) can be used to analyze continents and countries classified based on income so I will create an iso_code to directly detect them by using this trustable website https://www.countrycallingcodes.com/iso-country-codes

```python
# Continents
github.loc[github['country'] == 'Africa', 'iso_code'] = 'AF'
github.loc[github['country'] == 'Asia', 'iso_code'] = 'AS'
github.loc[github['country'] == 'Europe', 'iso_code'] = 'EU'
github.loc[github['country'] == 'North America', 'iso_code'] = 'NA'
github.loc[github['country'] == 'South America', 'iso_code'] = 'SA'
github.loc[github['country'] == 'Oceania', 'iso_code'] = 'OC'
github.loc[github['country'] == 'World', 'iso_code'] = 'WO'
```

```python
# Countries classified based on income
github.loc[github['country'] == 'High-income countries', 'iso_code'] = '111'
github.loc[github['country'] == 'Low-income countries', 'iso_code'] = '222'
github.loc[github['country'] == 'Lower-middle-income countries', 'iso_code'] = '333'
github.loc[github['country'] == 'Upper-middle-income countries', 'iso_code'] = '444'
```

```python
# Drop 'International transport' since it's not useful
github = github[~(github['country'] == 'International transport')]
```

```python
# Fix Kosovo and Timor
github.loc[github['country'] == 'Kosovo', 'iso_code'] = 'XXK'
github.loc[github['country'] == 'Timor', 'iso_code'] = 'TLS'
```

Next, we have to deal with columns of type float with NA values: Since they are float, I will substitute NA values by the mean of the column grouped by country so that I don't get false analysis: Imagine I substitute NA values with the minimum, I will have many numbers with minimum values in my plots and it may lead to false conclusions. Furthermore, I cannot let NA values in my data because it may cause some gaps in years (for example not having data for 2018 while having for 2017 and 2019), so mean is the best compromise.

```python
columns = list(information[information['Type'] == 'float64'].index)

for column in columns:
    github[column] = github.groupby('country')[column].transform(lambda x: x.fillna(x.mean()))

# Calculate the minimum value of each column
mean_values = github.select_dtypes(include=['float']).mean()

# Substitute NaN values with the minimum of each column
github = github.fillna(mean_values)
```

Let's check again the number of columns with NA values

```python
data_inf = {'Type':github.dtypes, 'Number_NAs':github.isna().sum()}
information = pd.DataFrame(data_inf)
information[information['Number_NAs'] > 0]
```

We are finally done with NAs

# Fossil CO2 emissions by country Dataset

The dataset provides information on CO2 emissions from different sectors such as Power Industry, Buildings, Transports... in various countries over several years. Each entry includes the sector, country code, country name, and CO2 emission values for each year from 1970 to 2021. The emissions are measured in metric tons and represent the contribution of each sector to CO2 emissions in each country. The data highlights the trends and changes in CO2 emissions from the power sector across different nations over time.
Source: https://edgar.jrc.ec.europa.eu/report_2022

**Why to combine it with CO2 dataset?** Merging the two datasets will help me analyze the contribution of different sectors to CO2 emissions, in addition to examining the relationship between fuel types (oil, gas, cement...) and the different sectors

```python
fossil = pd.read_excel('/content/fossil_CO2_booklet_2022.xlsx', sheet_name='fossil_CO2_by_sector_and_countr')
```

```python
fossil.head(3)
```

Let's see how many rows and columns we have

```python
fossil.shape
```

There is only CO2 as a Substance

```python
fossil['Substance'].unique()
```

So I can drop it

```python
fossil.drop('Substance', inplace = True, axis = 1)
```

I will rename the columns to cope with the CO2 dataset

```python
column_rename = {
    'EDGAR Country Code': 'iso_code',
    'Country': 'country',
    'Sector': 'sector_name'
}

fossil = fossil.rename(columns=column_rename)
```

I need also to convert the years from columns into rows

```python
# Convert years from columns to rows
fossil = fossil.melt(id_vars=['sector_name', 'iso_code', 'country'], var_name='year', value_name='fossil_co2_emission')
fossil.head(3)
```

Then, I need to have columns representing each fossil_co2_emission by sector, meaning that I want a column name  fossil_co2_emission_power_industry, fossil_co2_emission_buildings... depending on the values in sector name, and dropping sector_name column

**Why doing this?** Because when dealing with NA values, I want to average by the co2 emission by sector over the years not by the overall co2 emission over the years

```python
# Pivot the data to have columns representing each fossil CO2 emission by sector
fossil = fossil.pivot_table(index=['iso_code', 'country', 'year'], columns='sector_name', values='fossil_co2_emission')

# Reset the index to make 'iso_code', 'country', and 'year' as columns
fossil = fossil.reset_index()

# Rename the columns
fossil = fossil.rename(columns={
    'Power Industry': 'fossil_co2_emission_power_industry',
    'Buildings': 'fossil_co2_emission_buildings',
    'Other industrial combustion': 'fossil_co2_other_industrial_combustion',
    'Transport': 'fossil_co2_transport',
    'Other sectors': 'fossil_co2_other_sectors',
})

# Sort the columns in desired order
fossil = fossil[['iso_code', 'country', 'year', 'fossil_co2_emission_power_industry', 'fossil_co2_emission_buildings', 'fossil_co2_other_industrial_combustion','fossil_co2_transport','fossil_co2_other_sectors']]
```

```python
# I don't want the index to be called sector_name
fossil.columns.name = None

fossil.head(3)
```

Let's check again the shape

```python
fossil.shape
```

We need also to check NA values in parallel with the type of each column. For visualization purposes, I will display the type of each column next to number of NAs (in descending order)

```python
data_inf = {'Type':fossil.dtypes, 'Number_NAs':fossil.isna().sum()}
information = pd.DataFrame(data_inf)
information.sort_values(by = 'Number_NAs', ascending=False) # I am summming the NA values for each column and sorting them in descending order
```

Since they are float, I will substitute NA values by the mean of the column grouped by country so that I don't get false analysis: Imagine I substitute NA values with the minimum, I will have many numbers with minimum values in my plots and it may lead to false conclusions. Furthermore, I cannot let NA values in my data because it may cause some gaps in years (for example not having data for 2018 while having for 2017 and 2019), so mean is the best compromise.


```python
columns_to_fill = ['fossil_co2_emission_power_industry', 'fossil_co2_emission_buildings', 'fossil_co2_other_industrial_combustion', 'fossil_co2_transport', 'fossil_co2_other_sectors']

for column in columns_to_fill:
    fossil[column] = fossil.groupby('country')[column].transform(lambda x: x.fillna(x.mean()))

mean_values = fossil.select_dtypes(include=['float']).mean()
fossil = fossil.fillna(mean_values)
```

Let's have a look on the years this dataset covers

```python
years_fossil = sorted(list(fossil['year'].unique()))
print(years_fossil)
```

While the df dataset includes years from 1981 to 2021, this dataset  includes years from 1970 to 2021. So let's shrink the fossil dataset to 1981-2021 range


```python
fossil = fossil[(fossil['year'] > 1980)]
```

Let's look again at the shape

```python
fossil.shape
```

It's time to get the final merged dataset by combining the two dataframes by ISO code and year (I am not taking country because I found that there is a difference in how the country is represented in each of the dataframe)

```python
# Merge the datasets based on iso_code and year
df = pd.merge(df, fossil, on=['iso_code', 'year'], how='left')

# Drop the unnecessary columns
df = df.drop(['country_y'], axis=1)

# Rename country_x
df = df.rename(columns={'country_x': 'country'})
```

```python
# Rename iso_code
df = df.rename(columns={'iso_code': 'country_code'})
```

## CO2_emission_by_sector

```python
co2_emission_by_sector = df[['country_code', 'year', 'fossil_co2_emission_power_industry', 'fossil_co2_emission_buildings',
                             'fossil_co2_other_industrial_combustion', 'fossil_co2_transport', 'fossil_co2_other_sectors']]
```

```python
co2_emission_by_sector.head(3)
```

```python
# Define the sector names and sector codes
sector_mapping = {
    'fossil_co2_emission_power_industry': 0,
    'fossil_co2_emission_buildings': 1,
    'fossil_co2_other_industrial_combustion': 2,
    'fossil_co2_transport': 3,
    'fossil_co2_other_sectors': 4
}

# Split rows into sectors
sectors = [
    'fossil_co2_emission_power_industry',
    'fossil_co2_emission_buildings',
    'fossil_co2_other_industrial_combustion',
    'fossil_co2_transport',
    'fossil_co2_other_sectors'
]

# Split each row into five rows, where each row corresponds to one sector and map sector_name to sector_code
rows = []
for _, row in co2_emission_by_sector.iterrows():
    for sector in sectors:
        rows.append([row['country_code'],sector_mapping[sector], row['year'], row[sector]])

# Creating new DataFrame
co2_emission_by_sector = pd.DataFrame(rows, columns=['country_code', 'sector_code', 'year', 'co2_by_sector'])
```

## CO2_emission_by_fuel_type

```python
co2_emission_by_fuel_type = df[['country_code', 'year', 'cement_co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2', 'other_industry_co2',
                               'cement_co2_per_capita', 'coal_co2_per_capita', 'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita', 'other_co2_per_capita',
                               'cumulative_cement_co2', 'cumulative_coal_co2', 'cumulative_flaring_co2', 'cumulative_gas_co2', 'cumulative_oil_co2', 'cumulative_other_co2',
                               'share_global_cement_co2', 'share_global_coal_co2', 'share_global_flaring_co2', 'share_global_gas_co2', 'share_global_oil_co2', 'share_global_other_co2', 'share_global_cumulative_cement_co2', 'share_global_cumulative_coal_co2',
                               'share_global_cumulative_flaring_co2', 'share_global_cumulative_gas_co2', 'share_global_cumulative_oil_co2', 'share_global_cumulative_other_co2']]
```

```python
co2_emission_by_fuel_type.head(3)
```

```python
# Define the type names and type codes
type_mapping = {
    'cement': 0,
    'coal': 1,
    'gas': 2,
    'flaring': 3,
    'oil': 4,
    'other': 5
}

# Split rows into types
fuel_types = [
    'cement',
    'coal',
    'gas',
    'flaring',
    'oil',
    'other'
]

# For coding purposes
co2_emission_by_fuel_type = co2_emission_by_fuel_type.rename(columns={'other_industry_co2': 'other_co2'})

# Split each row into five rows, where each row corresponds to one type and map type_name to type_code
rows = []
for _, row in co2_emission_by_fuel_type.iterrows():
  for fuel_type in fuel_types:
      rows.append([row['country_code'],type_mapping[fuel_type], row['year'], row[fuel_type+'_co2'], row[fuel_type+'_co2_per_capita'],
                   row['cumulative_'+fuel_type+'_co2'], row['share_global_'+fuel_type+'_co2'], row['share_global_cumulative_'+fuel_type+'_co2']])

# Creating new DataFrame
co2_emission_by_fuel_type = pd.DataFrame(rows, columns=['country_code', 'type_code', 'year', 'co2', 'co2_per_capita', 'cumulative_co2',
                                                        'share_global_co2', 'share_global_cumulative_co2'])
```
